```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://survey.stackoverflow.co/2023/#technology-most-popular-technologies"
```

```python
reponse = requests.get(url)
```

```python
data = reponse.text
soup = BeautifulSoup(reponse.content, "html.parser")
```

```python
list_table = soup.find_all("table")
len(list_table)
```

```python
language = soup.find("table", {"id" : "language-prof5lolr"})
rows = language.find_all("tr")

Programming = pd.DataFrame(columns=["Index", "Language", "Percentage"])

for index, row in enumerate(rows):
    cols = row.find_all("td")
    name = cols[0].string
    id = index
    percent = cols[1].find_all("span")
    
    new_row = {"Index": id, "Language": name, "Percentage": percent[0].text}
    Programming = pd.concat([Programming, pd.DataFrame([new_row])], ignore_index=True)

Programming.drop("Index", axis = 1, inplace=True)
Programming
```

```python
database = soup.find("table", {"id":"database-learn4vrhw"})
rows1 = database.find_all("tr")

Databases = pd.DataFrame(columns=["Database", "Percentage"])

for row in rows1:
    cols = row.find_all("td")
    dtb = cols[0].string
    pct = cols[1].find("span").string
    new_row = {"Database": dtb, "Percentage":pct}
    Databases = pd.concat([Databases, pd.DataFrame([new_row])], ignore_index=True)
Databases
```
